```python
salary=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
y=salary['Salary']
x=salary[['Experience Years']]
```

```python
x.shape
```

```python
y.shape
```

```python
# step 4=train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529)
x_train
```

```python
#step5= which modle (liner regression)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#step 6=train the model(fit model)
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# y=c+mx(this is ml model)
#salary =26345+9398*exp+errors
```

```python
x_train
```

```python
x_test
```

```python
# step 7=prediction
```

```python
y_pred=model.predict(x_test)
```

```python
model.predict(x_train)
```

```python
# step 8= accuracy
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
